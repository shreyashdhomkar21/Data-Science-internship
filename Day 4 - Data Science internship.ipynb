```python
q3 = np.percentile(marks,75)

q1 = np.percentile(marks,50)
```

```python
iqr = q3-q1
iqr
```

## seaborn

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
import matplotlib.pyplot as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips, hue='time')
plt.show()
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips, hue='sex')
plt.show()
```

```python
#histogram

plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'] , kde=True)
plt.show()
```

```python
plt.title('Outliers in tip')
plt.xlabel('Day')
plt.ylabel('tip')
sns.boxplot(x='day' , y = 'tip' , data =tips)
plt.show()
```

```python
sns.pairplot(tips, hue='smoker');
plt.show()
```
