```python
#Feature Scaling-unit conversion,scale or spread same
#1.Normalization =MinMaxScaler ,range 0 to 1
#2.Standardization
```

```python
x=np.array([24,29,39,28,6,2,87,56,33])
```

```python
def norm(a):
    n1=(a-a.min())/(a.max()-a.min())
    return n1
norm(x)
```

```python
nx=norm(x)
```

```python
mp.figure(figsize=(16,9))

mp.subplot(1,2,1)
mp.title('Actual Data')
mp.plot(x)

mp.subplot(1,2,2)
mp.title('Normalized data')
mp.plot(nx)

#beacuse the max value is 1 and min value is 0 the y axis is then equally divided in range 0-1
```

```python
df1=pd.read_csv('Social_Network_Ads.csv')
df1
```

```python
x=df1[['Age','EstimatedSalary']]
y=df1['Purchased']
```

```python
mp.xlabel('age')
mp.ylabel('salary')
mp.grid()
mp.scatter(x['Age'],x['EstimatedSalary'],c=y)

#scatter(x, y, s, c, marker, cmap, norm, vmin, vmax, alpha, linewidths, edgecolors, plotnonfinite, data, **kwargs)
```

```python
x.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
#x_scaled is array
x_scaled=scaler.fit_transform(x)
type(x_scaled)
```

```python
x_new=pd.DataFrame(x_scaled,columns=x.columns)
x_new
```

```python
x_new.describe()
```

```python
mp.figure(figsize=(16,9))
mp.subplot(1,2,1)
mp.title('Acual Data')
mp.xlabel('age')
mp.ylabel('salary')
mp.grid()
mp.scatter(x['Age'],x['EstimatedSalary'],c=y)

mp.subplot(1,2,2)
mp.title('Normalize Data')
mp.xlabel('age')
mp.ylabel('salary')
mp.grid()
mp.scatter(x_new['Age'],x_new['EstimatedSalary'],c=y)
```
